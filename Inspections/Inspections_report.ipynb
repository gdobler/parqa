## I. Investigating Inspections time

- [Notebook](1_Inspections_time.ipynb)

Firts, we invistigated the corellation between inspection result and how much time inspectors spent.

![status_time](img/inspection_time.png)

- Same time, Neyther season, nor year, nor borough do not play any significant role in inspection time

## Notes

At the moment, while we failed to find any significant correlation between inspections and calls, we find interesting pattern of time spent on failing parks. Being significant on it's own, this may lead us to the question, why is that, was it the result of the rigorous inspection for places where something was wrong at the first place, or, for example. what we see is the result of larger area of the parks, which bring us back to the propensity in quality measurements
.

All this might bring us to new questions, some of them I listed below:
- Do the difference in time distribution for park inspections have strong correlation with property area
- Why there is specific pattern of inspections per park distribution for Brooklyn and Staten Island
- Why we can see negative correlation between number of inspections and percentage of failed ones and can we interpret that as inefficiency in selection of parks to inspect
- Can we improve inspection efficiency by reacting to number of calls for the park overpassing specific treshhold (might be different for each park)?
